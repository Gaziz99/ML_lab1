###  Classes
Classes provide a means of bundling data and functionality together. Creating a new class creates a new type of object, allowing new instances of that type to be made. Each class instance can have attributes attached to it for maintaining its state. Class instances can also have methods (defined by its class) for modifying its state. See https://docs.python.org/3/tutorial/classes.html for more information.


````python
class MyClass:
    
    # this is the constructor
    def __init__(self, a):
        self._a = a
     
    
    # this is a method
    def square(self, x):
        return x * x
    
    # this is a static method
    @staticmethod
    def mymethod(x):
        return x**3

mc = MyClass(3)  #Creates a new instance of the class 
print('method:', mc.square(4))  
print('static method:', MyClass.mymethod(5))
print('a:', mc._a)
````

## Autograd: Automatic Differentiation


A gradient measures how much the output of the given function varies when varying the inputs by a small factor, which is the same as the concept of derivatives in calculus. Gradients are the slope of a function. 

Autograd is a PyTorch package for the differentiation for all operations on Tensors. It performs the backpropagation starting from a variable and computes all the backpropagation gradients automatically. See https://colah.github.io/posts/2015-08-Backprop/ for more information.

In Pytorch, if we set a tensor's attribute `.requires_grad` as `True`, it starts to track all operations on it. When we finish our computation we can call `.backward()` and have all the gradients computed automatically. The gradient for this tensor will be accumulated into `.grad` attribute. Each tensor has a `.grad_fn` attribute that references a `Function` that has created the Tensor (except for Tensors created by the user - their `grad_fn` is `None`).

````python
# Create a tensor with no requires_grad
a = torch.ones(2, 2)

# Create a tensor and set requires_grad=True to track computation with it
x = torch.ones(2, 2, requires_grad=True)

print(a)
print(a.requires_grad)
print(a.grad_fn)

print(x)
print(x.requires_grad)
print(x.grad_fn)
````

````python
# Do a tensor opeartion on x
y = x + 2
print(y)
print(y.grad_fn)
````

````python
# Do more operations on y
z = y * y * 3
print(z)
print(z.grad_fn)
````

````python
z.backward(torch.tensor([[1., 1.],
                         [1., 1.]]))

print(x.grad)    #print gradients dz/dx when x = tensor([[1., 1.],
                 #                                       [1., 1.]])
````

One more example on scalar tensors

````python
# Create tensors.
x = torch.tensor(3., requires_grad=True)
w = torch.tensor(4., requires_grad=True)
b = torch.tensor(5., requires_grad=True)
print(x)
print(w)
print(b)
````

````python
# Arithmetic operations
y = w * x + b
y
````

`y` is a tensor with the value `4 * 3 + 5 = 17`. 

````python
# Compute derivatives
y.backward()
````

`y` contains a single scalar, `y.backward()` is equivalent to `y.backward(torch.tensor(1.))`.

````python
# Display gradients for y = w * x + b
print('dy/dx:', x.grad)   
print('dy/dw:', w.grad)
print('dy/db:', b.grad)
````

We can stop autograd from tracking history on Tensors with `.requires_grad=True` either by wrapping the code block in with `torch.no_grad()` or by using `.detach()` to get a new Tensor with the same content but that does not require gradients

````python
print(x.requires_grad)
print((x ** 2).requires_grad)

with torch.no_grad():                #stop tracking
    print((x ** 2).requires_grad)
````

````python
print(x.requires_grad)
y = x.detach()          #get a new Tensor with same content but does not require gradients
print(y.requires_grad)
print(x.eq(y).all())
````

## Neural Networks


A simple neural network architecture consists of three kinds of layer: the input layer, the output layer, and the hidden layer. When there is more than one hidden layer, it is called a deep neural network.
The following is a representation of a neural network:

<font size=1>image source:https://en.wikipedia.org/wiki/File:Colored_neural_network.svg </font>

In the preceding diagram the circles represent a neuron or in deep learning terms, a node, which is a computation unit. The edges represent the connection between the nodes and hold the connection weigh the two nodes. 

Neural networks can be constructed using the ``torch.nn`` package. An ``nn.Module`` contains layers, and a method ``forward(input)`` that returns the ``output``. See https://pytorch.org/docs/stable/nn.html#torch-nn-init for more information.

#### Define the network
Let us define a simple network with the following layer configurations: [784, 128,10]. This configuration represents the 784 nodes (28*28 pixels) in the input layer, 128 in the hidden layer, and 10 in the output layer. Inside the forward function, we will use the sigmoid activation function in the hidden layer.

Note: We use sigmoid activation function in this example, but there are a number of different activation functions that one could use in neural netowrks, each with its pros and cons. See https://pytorch.org/docs/stable/nn.html#non-linear-activations-weighted-sum-nonlinearity for more information.

````python
import torch
import torch.nn as nn

class MyFirstNet(nn.Module):                 # nn.Module subclass

    def __init__(self):                      # constructor
        super(MyFirstNet, self).__init__() 
        
        # a simple network with layer configurations: [784, 128,10].
        self.hidden = nn.Linear(784, 128)  # operation: y = Wx + b
        self.output = nn.Linear(128, 10)
  

    def forward(self, x):
        # define how the model is going to be run from input to output
        x = self.hidden(x)
        x = torch.sigmoid(x)
        x = self.output(x)
        return x


model = MyFirstNet()
print(model)
````

Let us try a random 128x784 input.

````python
input = torch.randn(128, 784)
out = model(input)
print(out)
````

#### Loss Function

A loss function takes the (output, target) pair of inputs, and computes a value that estimates how far away the output is from the target.

There are several different loss functions under the `nn` package . See https://pytorch.org/docs/stable/nn.html#loss-functions for more information. 

A simple loss is: `nn.MSELoss` which computes the mean-squared error between the input and the target. Here is an example:

````python
output = model(input)
target = torch.randn(1280)     # a dummy target, for example
target = target.view(128, -1)  # make it the same shape as output
criterion = nn.MSELoss()

loss = criterion(output, target)
print(loss)
````

Now, if we follow `loss` in the backward direction, using its `.grad_fn` attribute, we will see a graph of computations that looks like this:

input -> linear -> sigmoid -> linear -> MSELoss -> loss

````python
print(loss.grad_fn)  # MSELoss
print(loss.grad_fn.next_functions[0][0])  # Linear
print(loss.grad_fn.next_functions[0][0].next_functions[0][0])  # Sigmoid
````

#### Backprop

To backpropagate the error all we have to do is to `loss.backward()`. We need to clear the existing gradients though, else gradients will be accumulated to existing gradients.

Now we shall call `loss.backward()`, and have a look at the hidden layer's bias gradients before and after the backward.

````python
model.zero_grad()     # zeroes the gradient buffers of all parameters

print('hidden.bias.grad before backward')
print(model.hidden.bias.grad)

loss.backward()

print('hidden.bias.grad after backward')
print(model.hidden.bias.grad)
````

#### Update the weights

The simplest update rule used in practice is the Stochastic Gradient Descent (SGD):

`weight = weight - learning_rate * gradient`

We can implement this using simple Python code:

````python
learning_rate = 0.01
for f in model.parameters():
    f.data.sub_(f.grad.data * learning_rate)
````

However, as we use neural networks, we want to use various different update rules such as SGD, Nesterov-SGD, Adam, RMSProp, etc. To enable this, we a package called `torch.optim` that implements all these methods. Using it is very simple:

````python
import torch.optim as optim

# create your optimizer
optimizer = optim.SGD(model.parameters(), lr=0.01)  #lr is the learning rate

# in your training loop:
optimizer.zero_grad()   # zero the gradient buffers
output = model(input)
loss = criterion(output, target)
loss.backward()
optimizer.step()    # Does the update
````

### 1. Loading and normalizing CIFAR10

````python
import torch
import torchvision
import torchvision.transforms as transforms
from torch.utils.data.sampler import SubsetRandomSampler
import numpy as np
import matplotlib.pyplot as plt
````

````python
# The first transformation converts the raw data into tensor variables and 
# the second transformation performs normalization using the below operation:
#
#               x_normalized = x-mean / std
#
# The values 0.5 and 0.5 represent the mean and standard deviation for 3 channels: red, green, and blue.
#
#


transform = transforms.Compose(
    [transforms.ToTensor(),
     transforms.Normalize((0.5, 0.5, 0.5), (0.5, 0.5, 0.5))])
````

````python
# Download the train and test dataset (the first time)
trainset = torchvision.datasets.CIFAR10(root='./data', train=True, download=True, transform=transform)
testset = torchvision.datasets.CIFAR10(root='./data', train=False, download=True, transform=transform)
````

#### Training and Validation Datasets
While building real world machine learning models, it is quite common to split the dataset into 3 parts:
- Training set - used to train the model i.e. compute the loss and adjust the weights of the model using gradient descent.
- Validation set - used to evaluate the model while training, adjust hyperparameters (learning rate etc.) and pick the best version of the model.
- Test set - used to compare different models, or different types of modeling approaches, and report the final accuracy of the model. 

Since there's no predefined validation set in CIFAR10, we can set aside a small portion of the training set to be used as the validation set. 

````python
num_train = len(trainset)
indices = list(range(num_train))
split = int(np.floor(0.1 * num_train)) # split around 10% from the train set to validation set
np.random.seed(0)
np.random.shuffle(indices)

train_idx, valid_idx = indices[split:], indices[:split]
train_sampler = SubsetRandomSampler(train_idx)
valid_sampler = SubsetRandomSampler(valid_idx)
````

#### DataLoader
We can now created data loaders to help us load the data in batches. Large datasets requires loading them into memory all at once. This leads to memory outage and slowing down of programs. PyTorch offers a solution for parallelizing the data loading process with the support of automatic batching as well. This is the DataLoader class present within the `torch.utils.data` package.

````python
# Data loader
# automatically generates batches in the training loop and takes care of shuffling
# 

# load the trainset
trainloader = torch.utils.data.DataLoader(trainset, batch_size=4,sampler=train_sampler,
                                           num_workers=2)
# load the validset
validloader = torch.utils.data.DataLoader(trainset, batch_size=4,sampler=valid_sampler,
                                           num_workers=2)

# load the testset
testloader = torch.utils.data.DataLoader(testset, batch_size=4,
                                         shuffle=False, num_workers=2)

# Class labels
classes = ('plane', 'car', 'bird', 'cat', 'deer', 'dog', 'frog', 'horse', 'ship', 'truck')
````

We can look at batches of images from the dataset using the `make_grid` method from `torchvision`. Each time the following code is run, we get a different bach. 

````python
import matplotlib.pyplot as plt
import numpy as np

# functions to show an image
def imshow(img):
    img = img / 2 + 0.5     # unnormalize
    npimg = img.numpy()
    # in tensor, image is (batch, width, height), 
    # so you have to transpose it to (width, height, batch) in numpy to show it.
    plt.imshow(np.transpose(npimg, (1, 2, 0)))
    plt.show()


# get some random training images
dataiter = iter(trainloader)
images, labels = dataiter.next()  # trainloader batch_size=4, contains 4 images

#for i in range(4):
#    imshow(images[i])

# show the images
imshow(torchvision.utils.make_grid(images))

# print labels
for j in range(4):
    print(classes[labels[j]], end=" ")
````

### 2. Define a Neural Network

Let us define a neural network with different types of layers. See https://pytorch.org/docs/stable/nn.html for more information.


<font size=1>image source: https://i.imgur.com/KKtPOKE.png</font>

````python
import torch.nn as nn

class Net(nn.Module):
    def __init__(self):
        super(Net, self).__init__()
    
        # 3 input image channel, 6 output channels, 5x5 square convolution
        self.conv1 = nn.Conv2d(3, 6, 5)   
        # pooling over a (2, 2) window    
        self.pool = nn.MaxPool2d(2, 2)    
        # 6 input image channel, 16 output channels, 5x5 square convolution
        self.conv2 = nn.Conv2d(6, 16, 5)  
        # operation: y = Wx + b        
        self.fc1 = nn.Linear(16 * 5 * 5, 120)  # 5*5 from image dimension
        self.fc2 = nn.Linear(120, 84)
        self.fc3 = nn.Linear(84, 10)

    def forward(self, x):
        # define how the model is going to be run from input to output
        x = self.pool(torch.relu(self.conv1(x)))
        x = self.pool(torch.relu(self.conv2(x)))
        x = x.view(-1, 16 * 5 * 5)
        x = torch.relu(self.fc1(x))
        x = torch.relu(self.fc2(x))
        x = self.fc3(x)
        return x


model = Net() # create the model
````

### 3. Define a Loss function and optimizer

Let us use a Classification Cross-Entropy loss and SGD with momentum.

````python
import torch.optim as optim

criterion = nn.CrossEntropyLoss()
optimizer = optim.SGD(model.parameters(), lr=0.001, momentum=0.9)
````

### 4. Train the network

We simply need to loop over our data iterator, and feed the inputs to the network and optimize. 

An epoch is one single pass of all the input data through the network. Let us train the network for 5 passes over the training dataset, calculate and plot the training loss and validation loss.

````python
epochs = 5  # modify epochs to further train the network, say 10
train_losses = np.zeros(epochs)
valid_losses = np.zeros(epochs)

for epoch in range(epochs):  # loop over the dataset multiple times

    model.train()
    train_loss = []     
    for data in trainloader:
        # get the inputs; data is a list of [inputs, labels]
        inputs, labels = data

        # zero the parameter gradients
        optimizer.zero_grad()

        # forward propagation
        outputs = model(inputs)
        
        # loss calculation
        loss = criterion(outputs, labels)
        
        # backward propagation
        loss.backward()
        
        # weight optimization
        optimizer.step()

        train_loss.append(loss.item())
    
    model.eval()
    with torch.no_grad():
        valid_loss = []
        for data in validloader:
            # get the validset; data is a list of [inputs, labels]
            inputs, labels = data
       
            # forward propagation
            outputs = model(inputs)
        
            # loss calculation
            loss = criterion(outputs, labels)
        
            valid_loss.append(loss.item())
    
           
    # print statistics
    train_loss = np.mean(train_loss)
    valid_loss = np.mean(valid_loss)
    
    print( 'Epoch [%d] train loss: %.3f  valid loss: %.3f' %
         (epoch + 1, train_loss, valid_loss ))
    
    # save losses
    train_losses[epoch] = train_loss
    valid_losses[epoch] = valid_loss
print('Finished Training')
````

````python
# Plot the train loss per iteration
plt.plot(train_losses, label='train loss')
plt.plot(valid_losses, label='valid loss')
plt.legend()
plt.show()
````

Let us look at how the network performs on the whole dataset.



````python
correct = 0
total = 0
with torch.no_grad():
    for data in testloader:
        images, labels = data
        
        # forward pass
        outputs = model(images)
        
        # get prediction
        # torch.max returns both max and argmax
        # dim = 0 max along columns
        # dim = 1 max along rows
        _, predicted = torch.max(outputs.data, dim=1)
        
         # update counts
        total += labels.size(0)
        correct += (predicted == labels).sum().item()

print('Accuracy of the network on the test images: %d %%' % (
                                                   100 * correct / total))
````

It shows the network learnt something. The initial accuracy is around 10%, which is what one might expect from a randomly intialized model (since it has a 1 in 10 chance of getting a label right by guessing randomly).

Let us see what are the classes that performed well, and the classes that did not perform well:

````python
class_correct = list(0. for i in range(10))
class_total = list(0. for i in range(10))
with torch.no_grad():
    for data in testloader:
        images, labels = data
        outputs = model(images)
        _, predicted = torch.max(outputs, 1)
        c = (predicted == labels).squeeze()
        for i in range(4):
            label = labels[i]
            class_correct[label] += c[i].item()
            class_total[label] += 1


for i in range(10):
    print('Accuracy of %5s : %2d %%' % (
        classes[i], 100 * class_correct[i] / class_total[i]))
````

````python
# get all predictions in an array and plot confusion matrix
from sklearn.metrics import confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt

y_test = np.array(testset.targets)
p_test = np.array([])

with torch.no_grad():
    for data in testloader:
        images, labels = data

        # Forward pass
        outputs = model(images)

        # Get prediction
        # torch.max returns both max and argmax
        _, predictions = torch.max(outputs, 1)
  
        # update p_test
        p_test = np.concatenate((p_test, predictions.numpy()))

cm = confusion_matrix(y_test, p_test)
#print(cm)

plt.figure(figsize=(8, 6))
plt.title("Confusion Matrix")
ax = sns.heatmap(cm, annot=True, cmap="Blues", fmt='g')
ax.set(ylabel="True Label", xlabel="Predicted Label")
ax.set_xticklabels(classes)
ax.set_yticklabels(classes)
plt.show()
````

### 2. Custom nn Modules
We can modify the constructor of the neural netork model to accpet inputs to make the model more flexible. Here is an example

````python
import torch

class CustomNet(torch.nn.Module):
    def __init__(self, D_in, H, D_out):   # Constructor with input parameters
                                          # D_in: input dimension
                                          # H: hidden dimension
                                          # D_out: output dimension
        
        super(CustomNet, self).__init__()
        self.linear1 = torch.nn.Linear(D_in, H)
        self.linear2 = torch.nn.Linear(H, D_out)

    def forward(self, x):
       
        h_relu = self.linear1(x).clamp(min=0)
        y_pred = self.linear2(h_relu)
        return y_pred


# N is batch size; D_in is input dimension;
# H is hidden dimension; D_out is output dimension.
N, D_in, H, D_out = 64, 1000, 100, 10

# Create random Tensors to hold inputs and outputs
x = torch.randn(N, D_in)
y = torch.randn(N, D_out)

# Construct our model by instantiating the class defined above
model = CustomNet(D_in, H, D_out)

# Construct our loss function and an Optimizer. The call to model.parameters()
# in the SGD constructor will contain the learnable parameters of the two
# nn.Linear modules which are members of the model.
criterion = torch.nn.MSELoss(reduction='sum')
optimizer = torch.optim.SGD(model.parameters(), lr=1e-4)
for t in range(500):
    # Forward pass: Compute predicted y by passing x to the model
    y_pred = model(x)

    # Compute and print loss
    loss = criterion(y_pred, y)
    if t % 100 == 99:         # print the loss every 100 times
        print(t, loss.item())

    # Zero gradients, perform a backward pass, and update the weights.
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()
````
